### 한글 사용

```python
from matplotlib import font_manager, rc
import matplotlib as mpl
import platform

mpl.rcParams['axes.unicode_minus'] = False

system = platform.system()
if system == "Windows":
    font_path = "C:/Windows/Fonts/malgun.ttf"
elif system == "Darwin": # Mac OS
    font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
elif system == "Linux":
    font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
else:
    font_path = None

if font_path:
    font = font_manager.FontProperties(fname=font_path).get_name()
    mpl.rcParams['font.family'] = font
else:
    print("폰트 경로를 확인해주세요.")
```

## 차트

### Line

```python
# 그래프 크기설정 및 선 색상설정
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
t = np.arange(0.0, 5.0, 0.2)
title_font = {'family': 'Malgun Gothic', 'size': 16, 'color': 'blue', 'weight': 'bold'}
plt.figure(figsize=(10,10)) # 단위: 인치(가로,세로)
plt.subplot(3,2,1)
plt.title("Line chart example", fontdict=title_font, loc='right', pad=20)
plt.title("test", loc='left', pad=20, fontweight='bold')
plt.plot(x, y)
plt.subplot(3,2,2)
plt.plot(x, y, color='green', linestyle='dashdot')
plt.subplot(3,2,3)
plt.plot(x, y, color='red', linestyle='dashed', marker='o')
plt.subplot(3,2,4)
plt.plot(x, y, color='blue', linestyle='dashed', marker='o', markersize=12, markerfacecolor='red')
plt.subplot(3,2,5)
plt.title("chart style example")
plt.plot([10,20,30,40], [1,4,9,16], c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(0 ,50)
plt.ylim(-10, 30)
plt.subplot(3,2,6)
plt.plot(t, t, 'r--') # r(red), --(dashed)
plt.plot(t, 0.5 * t**2, 'bs:') # b(blue), s(square marker), :(dot)
plt.plot(t, 0.2 * t**3, 'g^-') # g(green), ^(triangle_up marker), -(solid)
plt.xticks([0,1,3,6], ['a','b','c','d'])
plt.yticks([0,1,4,9,16,32])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.tight_layout(pad=1.5)
plt.show()
```

```python
x = np.arange(0, 12, 0.1)
fig, ax = plt.subplots(2, 2)

ax[0,0].plot(x, np.sin(x))
ax[0,0].set_title('[0,0] sin')

ax[0,1].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[0,1].set_title('[0,1] sin & cos')
ax[0,1].legend(['sin', 'cos'])

ax[1,0].plot(x, np.tan(x))
ax[1,0].set_title('[1,0] tan')

ax[1,1].plot(x, np.exp(x))
ax[1,1].set_title('[1,1] exp')

plt.tight_layout(pad=1.5)
plt.show()
```

### Bar

```python
np.random.seed(0)
people = ['몽룡','춘향','방자','향단']
t = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
print("t", t)
print("performance", performance)

plt.subplot(1,2,1)
plt.title('bar chart (Vertical)')
plt.bar(t, performance)
plt.xticks(t, people)
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1,2,2)
plt.title('barh chart (Horizontal)')
plt.barh(t, performance, alpha=0.4) # 주의 : 첫번째 데이터 y가 y축으로 표현됨
plt.yticks(t, people)
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout(pad=1.5)
plt.show()
```

```python
# 데이터 프레임으로 바 그래프 그리기 1
df1 = pd.DataFrame({
    '나이':[15,20,17,50,2,30,23],
    '키':[165,150,151,15,80,175,185],
    '이름':['둘리','도우너','또치','길동','희동','마이클','영희']
}, columns=['나이','키','이름'])
# x = [i for i in range(len(df1['이름']))] # xticks 시 위치 표시에 사용할 변수
# x = list(df1.index)
print(df1)

df1.plot(kind='bar', grid=True)
plt.xticks(df1.index, df1['이름'], rotation='horizontal')
plt.show()

df1.plot(x='이름', y=['나이','키'], kind='bar', grid=True)
plt.show()

df1.plot(x='이름', y=['나이','키'], kind='barh', grid=True)
plt.show()

plt.bar(df1['이름'], df1['나이'])
plt.show()

df1[['나이','키']].plot(kind='bar', grid=True, figsize=(10,10))
plt.xticks(df1.index, df1['이름'], rotation='horizontal') # X축 이름 표시
plt.show()

df1[['나이','키']].plot(kind='barh', grid=True, figsize=(10,10))
plt.yticks(df1.index, df1['이름'], rotation='horizontal') # Y축 이름 표시
plt.show()
```

### Scatter

```python
count = 30
np.random.seed(0)
t = np.arange(count)
x = np.random.rand(count)
y = np.random.rand(count)
y1 = np.random.rand(count)
y2 = np.random.rand(count)
y3 = np.pi * (15 * np.random.rand(count))**2

plt.scatter(x, y1, c=y2, s=y3)
plt.show()

colormap = t
plt.scatter(x, y, s=50, c=colormap, marker='^')
plt.colorbar()
plt.show()

df = pd.DataFrame({'time':t, 'see':y})
df.plot('time', 'see', kind='scatter')
plt.show()
```

### Histogram

```python
np.random.seed(0)
x = np.random.randn(1000)
plt.hist(x)
plt.show()
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(10, 6))
plt.boxplot([s1,s2,s3])
plt.grid()
plt.show()
```

### Pie

```python
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
labels = ['개구리', '돼지', '개', '통나무']
explode = (0.03, 0.03, 0.05, 0.03)
size = [15, 30, 45, 10]

plt.figure(figsize=(10, 6))
plt.title('Pie Chart')
plt.pie(size, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```
