## 준비

```python
import pymysql as DBMS
import pandas as pd

def connect_db(database=""):
    try:
        db = DBMS.connect(host="localhost", port=3306, user="root", password="SeSAC_AI_3rd", charset="utf8", database=database)
        print(type(db), "connect", database)
        cursor = db.cursor()
        print(type(cursor))
        return db, cursor
    except Exception as e:
        print("[NG]", type(e), e)
        raise e

def show_databases(cursor):
    try:
        sql = "SHOW DATABASES"
        result = cursor.execute(sql)
        print(type(result), result, sql)
        result_tuple = cursor.fetchall()
        print(type(result_tuple))
        for row in result_tuple:
            print(len(row), row)
    except Exception as e:
        print("[NG]", type(e), e)
        raise e

def show_tables(cursor):
    try:
        sql = "SHOW TABLES"
        result = cursor.execute(sql)
        print(type(result), result, sql)
        result_tuple = cursor.fetchall()
        print(type(result_tuple))
        for row in result_tuple:
            print(len(row), row)
    except Exception as e:
        print("[NG]", type(e), e)
        raise e

def select_database(cursor):
    try:
        sql = "SELECT DATABASE()"
        result = cursor.execute(sql)
        print(type(result), result, sql)
        result_tuple = cursor.fetchall()
        print(type(result_tuple))
        for row in result_tuple:
            print(len(row), row)
    except Exception as e:
        print("[NG]", type(e), e)
        raise e

def use_database(cursor, database):
    try:
        sql = "USE " + database
        result = cursor.execute(sql)
        print(type(result), result, sql)
    except Exception as e:
        print("[NG]", type(e), e)
        raise e

def execute_sql(cursor, sql):
    try:
        result = cursor.execute(sql)
        print(type(result), result, sql)
    except Exception as e:
        print("[NG]", type(e), e, "\n", sql)
        raise e
```

### SHOW DATABASES

```python
db, cursor = connect_db()
show_databases(cursor)
db.close()
```

## SQL

### DROP DATABASE

```python
db, cursor = connect_db()
show_databases(cursor)
sql = "DROP DATABASE IF EXISTS ecommerce"
execute_sql(cursor, sql)
show_databases(cursor)
db.close()
print("close")
```

### CREATE DATABASE

```python
db, cursor = connect_db()
sql = "CREATE DATABASE IF NOT EXISTS ecommerce"
execute_sql(cursor, sql)
show_databases(cursor)
use_database(cursor, "ecommerce")
select_database(cursor)
db.close()
print("close")
```

### DROP TABLE

```python
db, cursor = connect_db("ecommerce")
show_tables(cursor)
sql = "DROP TABLE IF EXISTS {0}".format("product")
execute_sql(cursor, sql)
show_tables(cursor)
db.close()
print("close")
```

### CREATE TABLE

```python
sql = """
CREATE TABLE IF NOT EXISTS product (
PRODUCT_CODE VARCHAR(20) NOT NULL,
TITLE VARCHAR(200) NOT NULL,
ORI_PRICE INT,
DISCOUNT_PRICE INT,
DISCOUNT_PERCENT INT,
DELIVERY VARCHAR(2),
PRIMARY KEY(PRODUCT_CODE));
"""
db, cursor = connect_db("ecommerce")
show_tables(cursor)
execute_sql(cursor, sql)
show_tables(cursor)
db.close()
print("close")
```

### SELECT

```python
db, cursor = connect_db("ecommerce")
sql = "SELECT * FROM product"
execute_sql(cursor, sql)
result_tuple = cursor.fetchall()
print(type(result_tuple))
for row in result_tuple:
    print(row)
db.close()
```

### INSERT

```python
db, cursor = connect_db("ecommerce")
for i in range(3):
    product_code = 216573170 + i + 1
    sql = """
    INSERT INTO product (PRODUCT_CODE, TITLE, ORI_PRICE, DISCOUNT_PRICE, DISCOUNT_PERCENT, DELIVERY)
    VALUES ('%d', '%s', %d, %d, %d, '%s')""" % (product_code, "제목" + str(i), 10000, 8000, 20, "Y")
    execute_sql(cursor, sql)
db.commit()
db.close()
```

```python
db, cursor = connect_db("ecommerce")
sql = "UPDATE {table} SET {column}='{value}', DELIVERY='N' WHERE PRODUCT_CODE='{pk}'".format(table="product", column="title", value="test", pk="216573171")
execute_sql(cursor, sql)
db.commit()
db.close()
```
