```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=OPENAI_API_KEY)

# 오디오 파일 경로 정의
audio_file_path = os.path.abspath("Winston_Church.ogg")

# 사용할 모델 정의
model = "whisper-1"  # Whisper-1 모델 사용

# 오디오 파일을 바이너리 모드로 열기
with open(audio_file_path, 'rb') as audio_file:
    # 오디오 파일 전사(텍스트 변환)
    transcript = client.audio.transcriptions.create(
        model=model,
        file=audio_file
    )
    print(transcript.text)  # 전사된 텍스트 출력
```

```python
# 오디오 파일 경로 정의
audio_file_path = os.path.abspath("Winston_Church.ogg")

# 사용할 모델 정의
model = "whisper-1"  # Whisper-1 모델 사용

# 오디오 파일을 바이너리 모드로 열기
with open(audio_file_path, 'rb') as audio_file:
    # 오디오 파일 번역 및 텍스트 변환
    transcript = client.audio.translations.create(
        model=model,
        file=audio_file,
    )
    print(transcript.text)  # 번역된 텍스트 출력
```

```python
# 오디오 파일 경로 정의
audio_file_path = os.path.abspath(
    "cypher.mp3"  # 오디오 파일 경로를 설정합니다.
)

# 모델 정의
model = "whisper-1"  # Whisper-1 모델을 사용합니다.

# 프롬프트 정의
prompt = "Cypher"  # 올바른 철자를 힌트로 제공하기 위해 'Cypher'를 프롬프트로 설정합니다.

# 오디오 파일을 바이너리 모드로 열기
with open(audio_file_path, 'rb') as audio_file:
    # 오디오 파일 전사
    transcript = client.audio.transcriptions.create(
        model=model,
        file=audio_file,
        prompt=prompt
    )

# 전사된 텍스트 출력
print(transcript.text)
```

```python
# 오디오 파일 경로 정의
audio_file_path = os.path.abspath(
    "cypher.mp3"  # 오디오 파일의 경로를 설정합니다.
)

# 모델 정의
model = "whisper-1"  # Whisper-1 모델을 사용합니다.

# 오디오 파일을 바이너리 모드로 열기
with open(audio_file_path, 'rb') as audio_file:
    # 오디오 파일 전사(Transcription)
    transcript = client.audio.transcriptions.create(
        model=model,
        file=audio_file
    )

# GPT-4.1-mini를 사용하여 텍스트 후처리(post-processing)
system_prompt = """당신은 영화 전문가입니다.
오디오 파일의 전사를 수정하는 것이 당신의 역할입니다.
예를 들어, 'Cipher' 또는 'Cypher'의 단어는 'Cyber'로 수정해야 합니다."""

# 모델 정의
model = "gpt-4.1-mini"  # GPT-4o 모델을 사용합니다.

# 텍스트 후처리
response = client.chat.completions.create(
    model=model,
    messages=[
        {
            "role": "system",
            "content": system_prompt  # 시스템 프롬프트로 후처리 작업을 설정합니다.
        },
        {
            "role": "user",
            "content": transcript.text  # Whisper 전사 텍스트를 사용자 입력으로 전달합니다.
        }
    ]
)

# 수정된 텍스트 출력
print(response.choices[0].message.content)  # GPT-4.1-mini로 수정한 텍스트를 출력합니다.
```

## 무료(오픈 소스) 음성 인식 라이브러리 TOP 3

### 1. Faster-Whisper
**OpenAI의 Whisper 모델을 CTranslate2 엔진으로 다시 구현한 라이브러리**
- 특징: 원본 Whisper보다 최대 4배 빠르고 메모리 사용량이 훨씬 적습니다.
- 성능: 원본과 동일한 정확도를 유지하면서도 속도가 비약적으로 개선되어, 현재 개발자들이 가장 많이 사용하는 라이브러리입니다.
- 한국어: large-v3나 large-v3-turbo 모델을 사용하면 한국어 인식률이 매우 뛰어납니다.


```python
# !pip install faster-whisper
```

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "True"

from faster_whisper import WhisperModel
audio_file = "Winston_Church.ogg"

# 1. 모델 설정 (size: tiny, base, small, medium, large-v3, large-v3-turbo)
# 장치(device)는 "cuda"(GPU) 또는 "cpu"로 설정할 수 있습니다.
model_size = "large-v3-turbo"  # 한국어 인식률과 속도가 가장 균형 잡힌 모델
model = WhisperModel(model_size, device="cpu", compute_type="int8")

# 2. 음성 파일 변환 (transcribe)
# beam_size: 높을수록 정확도가 올라가지만 속도가 느려짐 (보통 5 권장)
segments, info = model.transcribe(audio_file, beam_size=5, language="ko")

print(f"감지된 언어: {info.language} (확률: {info.language_probability:.2f})")

# 3. 결과 출력 (segments는 제너레이터이므로 반복문으로 출력)
for segment in segments:
    print(f"[{segment.start:.2f}s -> {segment.end:.2f}s] {segment.text}")
```

### 2. Whisper-large-v3-turbo
**OpenAI에서 2024년 말에 공개한 최신 모델로, 속도 최적화에 중점을 둔 버전**
- 특징: 기존 large-v3 모델과 거의 대등한 정확도를 보여주면서 속도는 약 8배 가량 빠릅니다.
- 용도: 실시간에 가까운 변환이 필요하거나 저사양 환경에서 고성능을 내고 싶을 때 최적입니다.

### 3. Insanely-Fast-Whisper
**Hugging Face 팀에서 만든 라이브러리로, GPU 성능을 극한으로 끌어올려 사용**
