```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_classic.chat_models import ChatOpenAI
from langchain_classic.agents import load_tools, initialize_agent, AgentType

llm = ChatOpenAI(temperature=0, model_name='gpt-4.1-mini') # 창의성 0으로 설정
tools = load_tools(["wikipedia", "llm-math"], llm=llm) # llm-math의 경우 나이 계산을 위해 사용
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, description='계산이 필요할 때 사용', verbose=True)
agent.run("에드 시런이 태어난 해는? 2026년도 현재 에드 시런은 몇 살?")
```
