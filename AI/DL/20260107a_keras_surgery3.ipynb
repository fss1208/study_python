## 환경 준비

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

## 데이터 준비
- Diagnosis	1–3 (범주형)	진단 코드 (암/질환 타입 분류)
- FVC	실수값	Forced Vital Capacity — 노력성 폐활량
- FEV1	실수값	Forced Expiratory Volume — 1초간 호기량
- Performance Status	0–2	WHO 수행능력지수 (일상생활 가능 정도)
- Pain	T/F	흉통 여부
- Hemoptysis	T/F	객혈(피 섞인 가래) 여부
- Dyspnoea	T/F	호흡곤란 여부
- Cough	T/F	기침
- Weakness	T/F	허약감
- T-Stage	1–4	종양 크기·침윤 정도 (TNM stage)
- N-Stage	0–3	림프절 전이 (TNM stage)
- Diabetes Mellitus	T/F	당뇨병 여부
- MI Up To 6 Months	T/F	최근 6개월 내 심근경색 여부
- PAD	T/F	말초동맥질환 여부
- Smoking	T/F	흡연 여부
- Asthma	T/F	천식 여부
- Risk1Y (Label)	T/F	수술 후 1년 내 사망 → True, 생존 → False

```python
ds = np.loadtxt("ThoraricSurgery3.csv", delimiter=',')
x = ds[:, 0:16]
y = ds[:, 16]
print(type(ds), ds.shape)
print(type(x), x.shape)
print(type(y), y.shape)
```

## 구조 결정

```python
model = Sequential()
model.add(Dense(30, input_dim=16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## 모델 실행

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x, y, epochs=1, batch_size=16)
```

## 평가

```python
result_list = model.evaluate(x, y)
print(type(result_list), result_list)
print("Accuracy = %.4f" % (result_list[1]), "Loss = %.4f" % (result_list[0]))
```

# TEST

```python
def test_model(x, y, epochs, nodes):
    x_column_count = x.shape[1]
    model = Sequential()
    model.add(Dense(nodes, input_dim=x_column_count, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    history = model.fit(x, y, epochs=epochs, batch_size=x_column_count)
    result_list = model.evaluate(x, y)
    print("epochs = %d" % epochs, "nodes = %d" % nodes, ":","Accuracy = %.4f" % (result_list[1]), "Loss = %.4f" % (result_list[0]))
```

## epochs = [30, 60, 90], node = 30
- 30 : Accuracy = 0.8489 Loss = 0.3883
- 60 : Accuracy = 0.8404 Loss = 0.4088
- 90 : Accuracy = 0.8489 Loss = 0.3854

```python
epochs_list = [30, 60, 90]
for epochs in epochs_list:
    test_model(x, y, epochs, 30)
```

## node = [30, 60, 90], epochs = 30
- 30 : Accuracy = 0.8511 Loss = 0.3977
- 60 : Accuracy = 0.8511 Loss = 0.3909
- 90 : Accuracy = 0.8489 Loss = 0.3944

```python
nodes_list = [30, 60, 90]
for nodes in nodes_list:
    test_model(x, y, 30, nodes)
```
