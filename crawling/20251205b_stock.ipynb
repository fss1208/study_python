```python
import json
import urllib.request as req
from fake_useragent import UserAgent
ua = UserAgent()
```

## 하나의 종목정보 가져오기

```python
code = "A005930"
url = "https://finance.daum.net/api/quotes/" + code
headers = {'User-Agent': ua.chrome, 'referer': 'https://finance.daum.net/'}
res = req.urlopen(req.Request(url, headers=headers)).read().decode('utf-8')
stocks_json = json.loads(res)
stocks_json
```

## 일단위 주식가격 가져오기

```python
days = 10000
code = "A005930"
headers = {'User-Agent': ua.chrome, 'referer': 'https://finance.daum.net/'}
url = "https://finance.daum.net/api/quote/{0}/days?symbolCode={0}".format(code) # 30일
#url = "https://finance.daum.net/api/quote/{0}/days?symbolCode={0}&page=10&perPage=10&pagination=true".format(code) # 10일
#url = "https://finance.daum.net/api/quote/{0}/days?symbolCode={0}&perPage={1}&pagination=false".format(code, days) # x일 (최대:상장시작일)
res = req.urlopen(req.Request(url, headers=headers)).read().decode('utf-8')
stocks_json = json.loads(res)
len(stocks_json['data'])
#stocks_json['data']
for elmt in stocks_json['data']:
    print(elmt['date'], elmt['tradePrice'])
```

## 분단위 주식정보 가져오기

```python
days = 10000
code = "A005930"
headers = {'User-Agent': ua.chrome, 'referer': 'https://finance.daum.net/'}
url = "https://finance.daum.net/api/quote/{0}/times?symbolCode={0}".format(code) # 30분
#url = "https://finance.daum.net/api/quote/{0}/times?symbolCode={0}&page=1&perPage=10&pagination=true".format(code) # 10분
#url = "https://finance.daum.net/api/quote/{0}/times?symbolCode={0}&perPage={1}&pagination=false".format(code, days) # x분 (09:00 ~ 15:30)
res = req.urlopen(req.Request(url, headers=headers)).read().decode('utf-8')
stocks_json = json.loads(res)
len(stocks_json['data'])
#stocks_json['data']
for elmt in stocks_json['data']:
    print(elmt['date'], elmt['tradePrice'])
```

## 복수의 종목정보 한번에 가져오기

```python
headers = {'User-Agent': ua.chrome, 'referer': 'https://finance.daum.net/'}
url = "https://finance.daum.net/api/quotesv4?codes=KR7005930003,KR7005380001"
res = req.urlopen(req.Request(url, headers=headers)).read().decode('utf-8')
stocks_json = json.loads(res)['quotes']
#stocks_json
for elmt in stocks_json:
    print(type(elmt), elmt['name'], elmt['tradePrice'])
```

## 상위 10개 종목정보 가져오기

```python
headers = {'User-Agent': ua.chrome, 'referer': 'https://finance.daum.net/'}
url = "https://finance.daum.net/api/search/ranks?limit=10" # 다음 주식 요청 URL
res = req.urlopen(req.Request(url, headers=headers)).read().decode('utf-8')
rank_json = json.loads(res)['data']
#rank_json
for elmt in rank_json:
    print(type(elmt), elmt['rank'], elmt['name'], elmt['tradePrice'])
```
