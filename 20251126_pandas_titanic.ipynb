```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("titanic.csv")
del df["Unnamed: 0"]
df.head()
```

```python
df["sex"].value_counts(normalize=True)
```

```python
df["alive"].value_counts(normalize=True)
```

```python
temp = df[['sex','alive']].value_counts(normalize=True)
print(type(temp), '\n', temp) # Series
print(type(temp.index), '\n', temp.index) # MultiIndex
print(type(temp.values), '\n', temp.values) # numpy.ndarray
```

```python
s = df[['sex','alive']].value_counts(normalize=True)
print(type(s), '\n', s) # Series
asc = s.sort_values() # ascending=True
print(type(asc), '\n', asc) # Series
des = s.sort_values(ascending=False)
print(type(des), '\n', des) # Series
```
